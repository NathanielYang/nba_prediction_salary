### clean and remove rows where there is no player

```python
def cleaning(table_here, year):
    table_here.columns = table_here.columns.droplevel()
    table_here = table_here.query('Player != "Salary" and Player != "Player"')
    table_here.reset_index(inplace=True, drop=True)
    table_here['year'] = year
    return table_here
```
